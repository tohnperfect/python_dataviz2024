### import ข้อมูล indian director

```
data_director = pd.read_csv(os.path.join(path,'indian_director.csv'), encoding='latin-1')
data_director.head()
```

## ต่อแกน X


*   จับ 2 ตารางมาต่อกันเลย (merge)
*   เลือกมาเพิ่มเฉพาะบาง column (map)



```
data_director_sm = data_director[['Name',	'Known For',	'Birth Date']]
data_director_sm
```

```
data_india_sm = data_india[['Name',	'Year',	'Duration',	'Genre',	'Rating',	'Votes',	'Director']]
data_india_sm
```

```
# prompt: merge ตาราง data_director_sm กับ data_india_sm โดยเชื่อม Known For กับ Name

import pandas as pd
data_merge = pd.merge(data_director_sm,data_india_sm,left_on='Known For',right_on='Name')
data_merge
```

```
data_india[data_india['Name']=='Gangs of Wasseypur']
```

# ต่อแนวแกน X แบบ เลือกมาบาง column ด้วย map

```
data_director_sm['Known For'].map(data_india_sm.set_index('Name')['Duration'])
```

## ปัญหาเกิดเมื่อ ตารางรอง มีข้อมูลที่ Duplicate

```
data_india_sm[data_india_sm.duplicated(subset='Name',keep=False)] # check duplicate
```

```
data_india_sm[data_india_sm.duplicated(subset='Name',keep=False)]
```
